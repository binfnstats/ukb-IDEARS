```python
%load_ext autoreload
%autoreload 2

import os
import pandas as pd
import numpy as np
import ast
import sys

from pyarrow.parquet import ParquetFile
import pyarrow as pa 

ukb_file='../data/ukb50790.tab'
sys.path.append("../../ukb-dementia-shap/")

sys.path.append("../Pain/code/")



from logic.data_processing.data_import import dataload
from logic.data_processing.data_processing import data_proc_main

dp=data_proc_main()

def read_parq_nrow(file='../data/ukb_tp0.parquet',n=1000):
    pf = ParquetFile(file) 
    first_ten_rows = next(pf.iter_batches(batch_size = n)) 
    df = pa.Table.from_batches([first_ten_rows]).to_pandas() 
    return df


def exc_parquet(df,outfile='../data/ukb'):

    try:
        df.to_parquet(outfile+'.parquet')

    except:
        print('running exception')
        for c in df.columns:
            if df[df[c].apply(lambda x: isinstance(x, str))].shape[0]>0:
                df[c]=df[c].astype(str)
        df.to_parquet(outfile+'.parquet') 
        
    return df

def create_parquets():
    for i in range(9):

        print(i)
        if i<8:

            mini=i*1000
            maxi=(i+1)*1000-1
        if i==8:
            mini=i*1000
            rem=len(colsnew) % 8000
            maxi=8000+rem


        cols=list(set(['f.eid']+colsnew[mini:maxi]))
        df=pd.read_csv(ukb_file,sep='\t',usecols=cols)
        df.columns=[varmap[c] for c in df.columns]
        df=exc_parquet(df,outfile='../data/ukb'+str(i))

def load_parquets(file='../data/ukb',imin=0,imax=9):
    listof_df=[]

    for i in range(imin,imax):
        print(i)
        df=pd.read_parquet(file+str(i)+'.parquet')
        if i>0:
            df.drop(columns='eid',inplace=True)
        listof_df.append(df)
    df=pd.concat(listof_df,axis=1)
    
    return df

def sep_path(string,i=1):
    try:
        s=string.split('>')[i]
    except:
        s='N/A'
    return s
```

```python
def rename_vars(df=ukb_tp,cols=test_list):

    change_cols=[]
    for c in cols:


        if np.asarray(field_names2.loc[field_names2['col.name']==c,'Coding'])[0]>0 and df[c].nunique()<15 and 'date' not in c and 'device' not in c and df[c].count()>250000:

            try:
                ctemp=np.asarray(field_names2.loc[field_names2['col.name']==c,'Coding'].astype(int))[0]
                dic_temp=dict(np.asarray(code_mappings.loc[code_mappings['code']==ctemp,'dicts'])[0])


                df[c][pd.isnull(df[c])]=99
                df[c]=df[c].astype(int).astype(str)
                df[c][pd.notnull(df[c])]=df[c].map(dic_temp)
                change_cols.append(c)
            except:
                print(c)
                
    return df
            
```
