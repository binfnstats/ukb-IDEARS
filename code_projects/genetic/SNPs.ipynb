```python
files=[chrom1,chrom2,chrom3,chrom4,chrom6,chrom8,chrom10,chrom11,chrom12,chrom13,chrom14,chrom16]
def colrename(file,chrom=1,df=SNPlookup,var='Risk Variant'):
    mask_chrom=(df['Chromosome']==chrom)
    if file.shape[1]-6!=df[mask_chrom].shape[0]:
        print(str(chrom)+" fail")
        df_out=file[[m for m in range(file.shape[1]) if m==1 or m>5]]
        df_out.columns=np.append('eid',['chrom'+str(chrom)+str(m) for m in range(file.shape[1]) if m>5])
    else:
        print(str(chrom)+" success")
        df_out=file[[m for m in range(file.shape[1]) if m==1 or m>5]]
        df_out.columns=np.append('eid',[x for x in df[mask_chrom][var]])
        
    return df_out
    
```

```python
ADchrom1_rn=colrename(ADchrom1,1,SNPlookupAD,'SNP')
ADchrom2_rn=colrename(ADchrom2,2,SNPlookupAD,'SNP')
ADchrom5_rn=colrename(ADchrom5,5,SNPlookupAD,'SNP')
ADchrom7_rn=colrename(ADchrom7,7,SNPlookupAD,'SNP')
ADchrom8_rn=colrename(ADchrom8,8,SNPlookupAD,'SNP')
ADchrom11_rn=colrename(ADchrom11,11,SNPlookupAD,'SNP')
ADchrom14_rn=colrename(ADchrom14,14,SNPlookupAD,'SNP')
ADchrom18_rn=colrename(ADchrom18,18,SNPlookupAD,'SNP')
ADchrom19_rn=colrename(ADchrom19,19,SNPlookupAD,'SNP')
ADchrom20_rn=colrename(ADchrom20,20,SNPlookupAD,'SNP')
```

```python
chrom1_rn=colrename(chrom1,1)
chrom2_rn=colrename(chrom2,2)
chrom3_rn=colrename(chrom3,3)
chrom4_rn=colrename(chrom4,4)
chrom6_rn=colrename(chrom6,6)
chrom8_rn=colrename(chrom8,8)
chrom10_rn=colrename(chrom10,10)
chrom11_rn=colrename(chrom11,11)
chrom12_rn=colrename(chrom12,12)
chrom13_rn=colrename(chrom13,13)
chrom14_rn=colrename(chrom14,14)
chrom16_rn=colrename(chrom16,16)
```

```python
from functools import reduce
dfs=[chrom1_rn,chrom2_rn,chrom3_rn,chrom4_rn,chrom6_rn,chrom8_rn,chrom10_rn,chrom11_rn,chrom12_rn,
     chrom13_rn,chrom14_rn,chrom16_rn]
df_final = reduce(lambda left,right: pd.merge(left,right,on='eid'), dfs)
```

```python
ohe_cols=[col for col in df_final.columns if col!='eid']
df_ohe=pd.get_dummies(df_final[ohe_cols], prefix=ohe_cols)
df_ohe['eid']=df_final['eid']
```

```python
dfsAD=[ADchrom1_rn,ADchrom2_rn,ADchrom5_rn,ADchrom7_rn,ADchrom8_rn,ADchrom11_rn
       ,ADchrom14_rn,ADchrom18_rn,ADchrom19_rn,ADchrom20_rn]
df_final_AD = reduce(lambda left,right: pd.merge(left,right,on='eid'), dfsAD)
```

```python
ohe_cols_AD=[col for col in df_final_AD.columns if col!='eid']
df_ohe_AD=pd.get_dummies(df_final_AD[ohe_cols_AD], prefix=ohe_cols_AD)
df_ohe_AD['eid']=df_final_AD['eid']
```
